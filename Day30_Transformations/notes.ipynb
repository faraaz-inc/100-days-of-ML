# Variable Transformation

## Function Transformers

- We apple some mathematical formula on our data to transoform its values
- PDF of the data gets transoformed to ` Normal Distribution ` (or atleast close to it)
- In Machine Learning, statistical models work better with normally distributed data

Some Mathematical transformations:
- Log Transform
- Reciprocal Transform
- Box-Cox Transform
- Yeo Johnson statistics

### Function Transformer
Can be used for:
- Log transform
- Reciprocal transform
- Square-root transform
- Custom function transform

### Power Transform
Can be used for:
- Box-Cox transform
- Yeo Johnson

### How to determine if data is Normally Distributed?
- Call seaborns distplot function `sns.distplot()`
- Call Pandas skew function `pd.skew()`, if 0 - Good, else skewed
- QQ Plot - Most reliable

    

### Log Transform

- Take `log` of all the values
- Data gets closer to being normally distributed

**When to use Log Transform?**
- Can't be used on -ve values
- If data is right skewed, Log transform brings it to the center
    - Brings down large values to comparable scale with small values

### Other Transforms

#### Reciprical Transform (1/x)
- Big values become small and vice versa

#### Square transform
- Generally used for left skewed data
