# Batch Machine learning
- We train our model offline using complete data at once. There's no incremental learning.
- Once our model is trained, its deployed on servers.
- This process repeats periodically.

# Online Learning
- Training is done incrementally.
- Can be done in batches while the model is running in production.
- E.g., YouTube recommendations

### When to use Online Learning?
- Where there is a concept drift.
- Cost effective.
- Faster solution.

- Libraries for online learning:
    - River
    - Vowpal Wabbit

### Learning rate
- The pace at which your model changes (learns new stuff)

### Out of Core Learning
- When the data is so big you can't load all of it at once.
- So we use online learning and feed the data incrementally. 

### Disadvantages
- Tricky to use
- Risky
