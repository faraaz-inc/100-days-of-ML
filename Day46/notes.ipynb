# Curse of Dimensionality

Features -> Dimensions

- Only optimal no. of dimensions are useful, beyond a point, more no. of features can bring down the performance of a model

**Reason:** `Sparsity`

Problems:
- Performance decreases
- Computation increases

## Dimensionality Reduction

- Feature Selection
    - Forward selection
    - Backward elimination
    - Bi-directional elimination
- Feature extraction
    - Principle Component Analysis (PCA)
    - Linear Discriminant Analysis (LDA)
