# Instance Based Learning vs Model Based Learning

## Instance based learning
- Trying to find out similarities between query point and data instead of finding patterns in data.
- No training or learning.
- E.g.,
    - K-nearest neighbour
    - RBF
    - Kernel functions



## Model based learning
- Model tries to find patterns in the data and learn the behavior by training itself on the data.
- Model tries to find mathematical relations in the data to find decision points



- After training the model we don't need any data.
- Depending on the input and the decision function, output is given.
- E.g.,
    - Linear regression
    - Logistic regression
    - Decision trees
