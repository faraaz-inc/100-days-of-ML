## Power Transformer

### Box - Cox Transform



- The exponent here is a variable called `lambda` that varies over a range of `-5 to 5`
- In the process of searching, we examine all values of lambda
- Finally we choose the optimal value (resulting in best approximation to a normal distribution) for our variable

Internal working:
- Max. likelihood
- Bayesian statistics

Limitations:
- Only applicable to `numbers > 0`

### Yeo-Johnson transform

- Adjustment to Box-Cox transformation that allows us to apply it on negative numbers as well
