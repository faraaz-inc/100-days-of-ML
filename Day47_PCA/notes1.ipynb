# Principle Component analysis

**Benefits of PCA:**
- Brings down data from higher dimension to lower dimension, hence faster execution of algosithms
- Better visualization

**Feature selection:**
- In case we want to select one of two columns from a dataset, use the one that is having higher spread (variance)
- If, two columns have same variance, feature extraction comes to the rescue.
- Both columns can be used to extract a new feature from the existing features.

## Geometric intuition

- PCA finds a new set of coordinate axis
- In this new axis, we can choose the feature having higher variance
