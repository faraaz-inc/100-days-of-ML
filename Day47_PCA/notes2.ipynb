# PCA

- Each datapoint is projected on the new axis (unit vector)
- Hence, we get a scaler value for each projection (u.x)
- We'll choose the unit vector that gives max. variance

- This is the Mathematical objective function
- Using this expression, PCA will find the unit vector that gives us the max value of variance

### Covariance and Covariance matrix 

- Variance only tells the spread at one axis.
- Doesn't tell the relationship between two axes
- Covariance solves that problem

**Covariance matrix**

- Diagonal elements (variance) tell how data is spread across all axes
- Covariance elements tell the orientation of data

    ![image copy 11.png](<attachment:image copy 11.png>)

### Eigen vectors
- Vectors whose direction doesn't change on applying Linear transformations

### Eigen values
- stretching factor of eigen vector

    ![image copy 12.png](<attachment:image copy 12.png>)

- Applying a linear transformation on a vector is same as multiplying it by a scalar

**In PCA, the matrix used for Linear transformation is Covariance matrix**

- On solving the above mathematical objective, it turns out that the unit vector in the direction of highest variance (principle component) is the largest eigenvector of the covariance matrix.
- Hence, we need to perform Eigen decomposition of the Covariance matrix and choose the largest eigenvector.

## Step by Step solution

Consider a data initially in 3D space

## How to transform points?

**Transforming to 1D**
- Project the original data points on the transpose of principle component vector using dot product.

    ![image copy 14.png](<attachment:image copy 14.png>)

**Transforming to 2D**
- Same procedure, take dot product of transpose of eigen vector.
