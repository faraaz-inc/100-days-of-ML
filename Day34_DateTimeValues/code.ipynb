## Working with dates

```python
# Convert to datetime data type
date['date'] = pd.to_datetime(date['date'])
date.info()
```

```python
# Extract year
date['year'] = date['date'].dt.year

date.head()
```

```python
# Extract month
date['month'] = date['date'].dt.month
date.head()
```

```python
date['month_name'] = date['date'].dt.month_name()
date.head()
```

```python
# Extract days
date['date_day'] = date['date'].dt.day
date.head()
```

```python
date['day_name'] = date['date'].dt.day_name()
date.head()
```

```python
# Day of the week
date['date_dow'] = date['date'].dt.dayofweek

date.head()
```

```python
# Is weekend?
date['is_weekend'] = np.where(date['day_name'].isin(['Saturday', 'Sunday']), 1, 0)

date.head()
```

```python
# Extract week of the year
date['date_woy'] = date['date'].dt.isocalendar().week

date.head()
```

```python
# Extract quarter
date['quarter'] = date['date'].dt.quarter

date.head()
```

```python
# extract semester
date['semester'] = np.where(date['quarter'].isin([1, 2]), 1, 2)

date.head()
```

```python
# Time elapsed between two dates
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# month passed
np.round((today - date['date']).dt.days / 30.44, 0)
```
