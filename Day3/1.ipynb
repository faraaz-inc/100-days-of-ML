# Types of Machine Learning

- Supervised Learning
    - Regression
    - Classification
- Unsupervised Learning
    - Clustering
    - Dimensionality reduction
    - Anomaly detection
    - Association
- Semisupervised Learning
- Reinforcement Learning

## Supervised Learning

- If we have both input and output and the task is to form a relationship between them, then its called Supervised learning.

#### Regression
- If the output of a machine learning problem is numerical, then its called Regression

#### Classification
- If the output of a machine learning problem is categorical (Yes/No), then its called Classification.

## Unsupervised Learning
- When there's only input data and no associated output data.

#### Clustering
- We can classify data in different clusters. Useful for further processing and interpretation of data.

#### Dimensionality reduction
- Used to convert multiple related dimensions (columns) of the data into a single one.(Feature extraction)

#### Anomaly Detection
- Used to detect irregularities in data

#### Association rule based learning
- Used to find hidden relationships (associativity) among data.

## Semisupervised Learning

- Instead of labelling all the data points for supervised learning, we only label a few. Rest are automatically generated by semi supervised models.
- E.g., Face recognition in Google Photos

## Reinforcement Learning
- No data is given initially to the model.
- Algorithm starts from scratch and learns gradually
- Similar to training a dog, give a treat for good stuff, punish otherwise.
