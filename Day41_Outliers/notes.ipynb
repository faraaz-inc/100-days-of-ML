# Outliers

- Can distract the ML models from actual data

## Effect of Outliers

- Linear Regression, Logistic Regression, Adaboost, Deep Learning are all affected negatively by outliers
- If the algorithm uses weight calculation, then it will be affected by outliers

## How to treat outliers?

- Trimming
    - Causes thinning of data
- Capping
- Treat as missing value
- Discretization

## How to detect Outliers?

**Normally distributed**
- If the data is normally distributed, then if a datapoint is outside +-3 std. deviations from mean, then it is an Outlier
