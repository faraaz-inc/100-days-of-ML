# Outlier removal using Z-Score

- Can be applied in case data is normally distributed
- In case data is outside +- 3 std. deviation, then it can be treated as outlier

![image copy 2.png](<attachment:image copy 2.png>)

## Outlier treatment

**Trimming**
- Remove the outliers
- If outliers are large in no. then it can reduce the dataset significantly

**Capping**
- Change outlier values to the +- 3 st.deviation values
